```python
import pandas as pd
import os
import re
file_path="C:\\Users\\bedlamini\\Documents\\VAT Schedules\\RES Files\\NEW RES -ERS Sales Purchase VAT MSCO.xlsx"
df=pd.read_excel(file_path,sheet_name=["INPUT TAX","OUTPUT_TAX"], header=0)

input_df=df.get("INPUT TAX")
output_df=df.get("OUTPUT_TAX")
input_df['index_column']=input_df.index
output_df['index_column']=output_df.index
```

```python
output_df.shape
```

```python
output_df_pv=output_df.pivot(index=['index_column','Document Date','Customer Name','Customer VAT Registration Number','Reference','Tax BaseAmount'], 
     columns='Tax Code Description',values=['Tax Amount']).stack(level=0).reset_index()
output_df_pv.head(10)

    
```

```python
output_df_pv=output_df_pv.assign(zeroexports='',Adjustment='',Adjustmentother='',DescriptionGood='')
output_df_pv = output_df_pv.fillna(value=0)
output_df_pv['VAT']=output_df_pv['0% Output Tax'] + output_df_pv['15% Output VAT']
output_df_pv.head(5)
```

```python
column_order=['Document Date','Customer VAT Registration Number','Customer Name','Reference','DescriptionGood','Output Tax Exempt','zeroexports','0% Output Tax','15% Output VAT','Adjustment','Adjustmentother','VAT','Tax BaseAmount']
df=output_df_pv.reindex(column_order,axis=1)
df.head(2)
```

```python
df.head(10)
```

```python
with pd.ExcelWriter('C:\\Users\\bedlamini\\Documents\\VAT Schedules\\RES Files\\New VAT Schedule___Updated.xlsx',engine='openpyxl',mode='a',if_sheet_exists='overlay') as writer:
    df.to_excel(writer,sheet_name='VAT schedule form--output tab',startrow=2,header=False,index=False)
```
